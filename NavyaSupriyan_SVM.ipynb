```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.figure_factory as ff
import plotly.express as px
from sklearn import metrics
from sklearn.metrics import confusion_matrix, precision_recall_curve, auc, roc_curve, accuracy_score, recall_score, classification_report, f1_score, average_precision_score, precision_recall_fscore_support, roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv('/content/drive/MyDrive/IRIS.csv')
```

```python
train_data,test_data=train_test_split(df,random_state=0)
```

```python
x=train_data.drop(['species'],axis=1)
y=train_data.species
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
x_test
```

```python
y_test
```

```python
models={
 "Support Vector Machine":SVC(), "K Nearest Neighbor":KNeighborsClassifier(weights='distance',metric='euclidean',n_neighbors=1), "Logistic Regression": 
 LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True, intercept_scaling=1, l1_ratio=None, max_iter=100, multi_class='auto', 
                          n_jobs=None, penalty='l2',random_state=0, solver='lbfgs', tol=0.0001, verbose=0, warm_start=False)
}
```

```python
for name, model in models.items():
  model.fit(x_train,y_train)
  print(name + ' trained.')
```

```python
original_results = []
data = []
columns = ["Accuracy", "Precision", "Recall", "F1-Score"]

for name, model in models.items():
  result = model.score(x_test, y_test)
  original_results.append(result)

  print("""
  __________________________"""+name+"""__________________________
  """)

  model = models[name]
  y_test_pred = model.predict(x_test)
  arg_test = {'y_true':y_test, 'y_pred':y_test_pred}
  
  accuracy = model.score(x_test, y_test)
  precision = metrics.precision_score(y_test, y_test_pred, average='macro')
  recall = metrics.recall_score(y_test, y_test_pred, average='macro')
  f1_score = metrics.f1_score(y_test_pred, y_test, average='macro')

  values = [accuracy, precision, recall, f1_score]
  zipped = zip(columns, values)
  a_dictionary = dict(zipped)
  ##print(a_dictionary)
  data.append(a_dictionary)

  print(confusion_matrix(**arg_test))
  print(classification_report(**arg_test))

  print(name + ': {:.5f}%'.format(result * 100))
```
